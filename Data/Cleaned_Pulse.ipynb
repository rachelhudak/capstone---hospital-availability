```python
import pandas as pd

# Skip the metadata rows
df_income = pd.read_csv(
    "HDPulse_data_export.csv",
    skiprows=4
)

# Clean income column
df_income["Value (Dollars)"] = (
    df_income["Value (Dollars)"]
    .str.replace(",", "", regex=False)
    .astype(float)
)

df_income.head()
```
