```python
import pandas as pd

df = pd.read_csv("Hospital_General_Information.csv")

df.head()
```

```python
df.info()
```

```python
df = df.replace("Not Available", pd.NA)


df["Hospital overall rating"] = pd.to_numeric(
    df["Hospital overall rating"], errors="coerce"
)

for col in df.columns:
    if "Count" in col or "Measure" in col:
        df[col] = pd.to_numeric(df[col], errors="coerce")

df.info()
```

```python
df_rating = df.dropna(subset=["Hospital overall rating"])
```
